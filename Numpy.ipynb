## NumPy Views vs. Copies

When working with NumPy arrays, it's important to understand the distinction between a **view** and a **copy**.

*   **View**: A view is a *reference* to the original array's data. It's like looking at the same data through a different window (with potentially a different shape or selection of elements). **Changes made to a view will directly affect the original array.** This is because both the view and the original array share the same underlying data buffer in memory.

*   **Copy**: A copy is a *new* array that contains a complete duplicate of the original array's data. It has its own independent memory location. **Changes made to a copy will NOT affect the original array.**

In your notebook, the code demonstrates this:

*   `view = arr[1:3]` creates a **view**. When you modify `view[1]`, the corresponding element in the original `arr` is also changed.
*   `copy = arr[1:3].copy()` creates a **copy**. When you modify `copy[1]`, the original `arr` remains unchanged.

Understanding this difference is crucial for avoiding unexpected behavior when manipulating NumPy arrays.

#Broad Casting
Broad casting is automatically stretch the smaller array to match the shape of the bigger array.
Rules(very important):Numpy compares the shape of the array from the end(right to left)
1)If shapes are equal,they are compatible.
2)If 1 is 1,it can be stretched to match the other.
3)If shapes are different and not 1 or equal then ERROR

Broadcasting is a powerful feature in NumPy that allows you to perform operations on arrays of different shapes. When NumPy performs an operation (like addition, subtraction, multiplication, etc.) on two arrays, it checks if they are compatible according to a set of rules. If they are, NumPy will "stretch" or "broadcast" the smaller array to match the shape of the larger array so the operation can be performed element-wise. This avoids the need to explicitly create larger temporary arrays, which can save memory and improve performance.

Here are the rules NumPy follows when comparing the shapes of two arrays, starting from the trailing (rightmost) dimension:

Equal dimensions: If the dimensions are the same size, they are compatible.
One dimension is 1: If one of the dimensions is 1, it can be stretched (broadcast) to match the size of the other dimension.
Dimensions are different and neither is 1: If the dimensions are different and neither is 1, the arrays are incompatible, and a ValueError will be raised.

```python
#Broad Casting
image=np.array([[200,150],[100,250]])
print(image)
```

```python
brightness=image+50
print(brightness)
```

#vectorization
#The np.vectorize() function is used to create a vectorized function from a regular Python function. This can be helpful for applying a function that is not already vectorized by NumPy to elements of a NumPy array. It's important to note that np.vectorize() is primarily for convenience and does not necessarily improve performance compared to a carefully implemented NumPy vectorized operation.

```python
#np.vectorize()->Convert a regular function to be applied on an array.
def sqrt(x):
  return x*x
vfunc=np.vectorize(sqrt)
arr=np.array([1,2,3])
print(vfunc(arr))
```

#vectorized(pros)
